```python
text = input("enter a sentence: ").lower().split()
words = {}
for word in text:
    words[word] = words.get(word,0) + 1
for word in words:
    print(f"the word '{word}' has appeard {words.get(word)} times")
```

```python
sum = 0
menu = {"hamburger" : 50,
        "salad" : 15,
        "cheese" : 20,
        "pizza" : 35
        }
print("what would you like to order from the menu?\n")
for item in menu:
    print(f"{item} with the price of {menu.get(item)}")
order = input("enter your order here: ").lower().split()
orderDic = {}
for item in order:
    orderDic[item] = orderDic.get(item,0) + 1
for item in orderDic:
    sum += menu.get(item,0) * orderDic.get(item)
print(f"your total bill is {sum}")
```

```python
students = {"student1" : {"name" : "neal","grades" : [99,87,90,94],"attendance":80},
            "student2": {"name" : "brody", "grades" : [80,85,79,80],"attendance" : 70},
            "student3" : {"name" : "alex", "grades" : [70,73,80,89], "attendance" : 90}}
for student in students:
    print(students[student]["name"],"got the average score of", sum(students[student]["grades"]) / len(students[student]["grades"]))
bestAttendence = 0
bestStudent = ""
for student in students:
    if students[student]["attendance"] > bestAttendence:
        bestAttendence = students[student]["attendance"]
        bestStudent = students[student]["name"]
print(f"{bestStudent} got the best attendance precentage with {bestAttendence}%")

```
