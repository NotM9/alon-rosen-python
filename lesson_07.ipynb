```python
def celsius_to_fahrenheit(celsius):
    return celsius* 9/5 + 32
print(celsius_to_fahrenheit(0))
print(celsius_to_fahrenheit(25))
print(celsius_to_fahrenheit(100))
```

```python
def get_grade(score):
    if score < 60:
        print('F')
    elif score < 70:
        print('D')
    elif score < 80:
        print('C')
    elif score < 90:
        print('B')
    else:
        print('A')
get_grade(95)
get_grade(73)
get_grade(80)
```

```python
def analyze_list(numbers):
    value = sum(numbers)
    length = len(numbers)
    average = value/length
    return value,length,average
numbers = {2,4,6}
stats = analyze_list(numbers)
print(stats)
```

```python
def add(a,b):
    return a +b
def subtract (a,b):
    return a-b
def multiply(a,b):
    return a*b
def divide(a,b):
    if b == 0:
        return 0
    return a/b
def calculate():
    userInput = input("Welcome to the calc\n what do your want to do?\n (a)dd   (s)ubtract  (m)ultiply  (d)ivide\n")
    a = (int) (input("write the first number: "))
    b = (int) (input("write the second number: "))
    match userInput.lower():
        case "a": resualt = add(a,b)
        case "s": resualt = subtract(a,b)
        case "m": resualt = multiply(a,b)
        case "d": resualt = divide(a,b)
    return resualt
print(calculate())
```
