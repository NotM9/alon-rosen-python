```python
foods = ["pizza", "burger", "sushi", "pasta", "salad"]
counter = 0

for food in foods:
    print(food)
    counter += 1

print("Total number of foods:", counter)
```

```python
numbers = [10, 20, 30, 40, 50]
sum = 0
counter = 0
for number in numbers:
    sum += number
    counter += 1
average = sum / counter
print("Sum:", sum)
print("Average:", average)
```

```python
shopping_list = []
for i in range(5):
    newItem = input("Enter an item to add to your shopping list: ")
    shopping_list.append(newItem)
for item in shopping_list:
    print(item)
```

```python
numbers = [23, 67, 12, 89, 45, 34]
maximum = numbers[0]
for number in numbers:
    if number > maximum:
        maximum = number
print("The largest number is:", maximum)
```
