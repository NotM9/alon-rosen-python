```python
num = int(input("Enter a number: "))
i = 1
while i <= 10:
    print(num*i)
    i += 1
```

```python
sum = 0
repetitions = 0
while True:
    num = int(input("Enter number: "))
    if num == -1:
        if repetitions == 0:
            print(0)
        else:
            print(sum/repetitions)
        break
    if num>=0:
        sum += num
        repetitions += 1
        continue
    print("negative numbers aren't allowed")
```

```python
import random
points = 100
score = 0
print("Welcome to the game\n press enter to move to next round")
while points > 0:
    input()
    bonus = random.randint(1,5)
    points -= random.randint(1,15)
    score += bonus
    print(f"you got {bonus} points added to your score")
print(f"the game has finished, you erned {score} points")
```
